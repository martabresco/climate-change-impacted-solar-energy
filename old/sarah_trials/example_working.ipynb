```python
rsds_ds=ds['SID']
rsds_ds2=ds2['influx_direct']
rsdsdiff_ds2=ds2['influx_diffuse']
```

```python
# Convert cftime time in tas to datetime64 (if needed)
if not np.issubdtype(rsds_ds['time'].dtype, np.datetime64):
    rsds_ds = rsds_ds.assign_coords(time=rsds_ds.indexes['time'].to_datetimeindex())

# Convert cftime time in tas to datetime64 (if needed)
if not np.issubdtype(rsds_ds2['time'].dtype, np.datetime64):
    rsds_ds2 = rsds_ds2.assign_coords(time=rsds_ds2.indexes['time'].to_datetimeindex())
```

```python
specific_time= '2013-01-01T01:00:00'
rsds_ds_time=rsds_ds.sel(time=specific_time, method='nearest')
rsds_ds2_time=rsds_ds2.sel(time=specific_time, method='nearest')
```

```python
lon_min, lon_max = rsds_ds_time.lon.min().item(), rsds_ds_time.lon.max().item()
lat_min, lat_max = rsds_ds_time.lat.min().item(), rsds_ds_time.lat.max().item()
# Create the plot
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the solar irradiance at the selected time
rsds_ds_time.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap='coolwarm', # Colormap for solar irradiance
    cbar_kwargs={'label': 'Solar direct irradiance (W/m^2) data CMSAF'}
)

# Add map features for context

ax.coastlines(resolution='10m', linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='white')

# Set extent to Europe
# Set extent to match the atlite cutout boundaries
ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())


# Add title and labels
plt.title(f'Average Solar Direct Irradiance over Europe at {specific_time}, data CMSAF')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
