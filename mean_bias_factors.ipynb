```python
import os
def calculate_mean_nc(filename):
    try:
        # Open the .nc file using xarray
        ds = xr.open_dataset(filename)
        if "bias_factor" in ds:
            # Calculate the mean of the bias_factor variable
            mean_value = ds.bias_factor.median().item()  # Convert to a Python float
            return mean_value
        else:
            return "bias_factor variable not found"
    except FileNotFoundError:
        return "File not found"
    except Exception as e:
        return f"Error processing file: {e}"

def process_folder_nc(folder_path):
    if not os.path.isdir(folder_path):
        print("Invalid folder path.")
        return

    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path) and file_name.endswith(".nc"):  # Ensure it's a .nc file
            logging.info(f"Processing file: {file_path}")
            mean_value = calculate_mean_nc(file_path)
            print(f"{file_name}: {mean_value}")

# Example usage
folder = "/work/users/s233224/Climate-Change-Impacted-Solar-Energy-Generation/bias_factors/"
process_folder_nc(folder)
```
